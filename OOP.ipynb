Constructor y self
- EL constructor es una función especial llamda __init__
- Acepta por parámetro una variable llamada self, que apunta a la instancia (objeto) de la clase

```python
class Mascota():
    def __init__(self, chip, nombre):
        #variables de instancia
        self.id_chip = chip ## a self es porque se está apuntando a las variables de la instancia
        self.nombre = nombre

mascota1 = Mascota(1,"Sauron")
mascota2 = Mascota(2,"Rex")

print(f"Mascota de chip {mascota1.id_chip}, su nombre es {mascota1.nombre}")
print(f"Mascota de chip {mascota2.id_chip}, su nombre es {mascota2.nombre}")
```

Método de instancia
- Se crea un método para describir a la instancia. Es más ordenado, simples y sencillo.
- Se crea como una función
- En este caso se refiere al método descripción_mascota

```python
class Mascota():
    def __init__(self, id_chip, nombre):
        self.id_chip = id_chip
        self.nombre = nombre
    def descripcion_mascota(self): # Self para tener acceso a variables de instancia
        print(f"Mascota de chip {self.id_chip}, su nombre es {self.nombre}")
mascota1 = Mascota(1,"Sauron") #instancia u objeto (mascota1)
mascota2 = Mascota(2,"Rex")        
mascota1.descripcion_mascota()
mascota2.descripcion_mascota()
```

Variable de clase
- Variable a nivel de clase no de objeto
- Asociada a toda la clase y a ninguna instancia en particular

```python
class Mascota():
    #Variable de clase
    descuento = 0.1
    def __init__(self, id_chip, nombre, precio):
        self.id_chip = id_chip
        self.nombre = nombre
        self.precio = precio
    def descripcion_mascota(self): # Self para tener acceso a variables de instancia #Método de instancia
        print(f"Mascota de chip {self.id_chip}, su nombre es {self.nombre}")
    def aplicar_descuento(self):
        self.precio = int(self.precio *(1-self.descuento))
        return self.precio
    
mascota1 = Mascota(1,"Sauron",1000)
mascota2 = Mascota(2,"Rex",1200)

Mascota.descuento = 0.2
print(f"Antes de descuento el valor de {mascota1.nombre} es de {mascota1.precio}")
mascota1.aplicar_descuento()
print(f"Después de descuento el valor de {mascota1.nombre} es de {mascota1.precio}")

#valor = mascota1.aplicar_descuento()
#print(valor)

mascota2.aplicar_descuento()
print(mascota2.precio)
```

Métodos de clase

```python
class Mascota():
    #Variable de clase
    descuento = 0.1
    def __init__(self, id_chip, nombre, precio):
        self.id_chip = id_chip
        self.nombre = nombre
        self.precio = precio
    def descripcion_mascota(self): # Self para tener acceso a variables de instancia #Método de instancia
        print(f"Mascota de chip {self.id_chip}, su nombre es {self.nombre}")
    def aplicar_descuento(self):
        self.precio = int(self.precio *(1-self.descuento))
        return self.precio
    @classmethod
    def modificar_descuento(cls, nuevo_descuento):
        cls.descuento = nuevo_descuento

mascota1 = Mascota(1,"Sauron",1000)
mascota2 = Mascota(2,"Rex",1200)
print(mascota1.precio)
print(mascota2.precio)
Mascota.modificar_descuento(0.5)
mascota1.aplicar_descuento()
mascota2.aplicar_descuento()
print(mascota1.precio)
print(mascota2.precio)
```

Métodos estáticos
- No tiene relación con la clase ni la instancia
- Es un método utilitario. No modifica elementos de clase o instancia

```python
class Mascota():
    #Variable de clase
    descuento = 0.1
    def __init__(self, id_chip, nombre, precio):
        self.id_chip = id_chip
        self.nombre = nombre
        self.precio = precio
    def descripcion_mascota(self): # Self para tener acceso a variables de instancia #Método de instancia
        print(f"Mascota de chip {self.id_chip}, su nombre es {self.nombre}")
    def aplicar_descuento(self):
        self.precio = int(self.precio *(1-self.descuento))
        return self.precio
    @classmethod
    def modificar_descuento(cls, nuevo_descuento):
        cls.descuento = nuevo_descuento
    @staticmethod
    def llamar_veterinario(numero):
        print(f"LLamando al numero {numero}")

mascota1 = Mascota(12,"Sauron",1500)
mascota2 = Mascota(21,"Rex",1400)
mascota1.llamar_veterinario(155848)
Mascota.llamar_veterinario(8558584) #Se puede llamar desde la clase o la instancia
```

Herencia
- Una clase que hereda atributos y metodos de otra clase

```python
class Perro(Mascota):
    def ladrar(self):
        print(f"Guau, mi nombre es {self.nombre}, guau")
    pass
perro1 = Perro(125,"Canito",1450)
perro1.ladrar()
```

```python
class Perro(Mascota):
    def __init__(self, id_chip, nombre, precio, raza):
        super().__init__(id_chip, nombre, precio) # A través de super se llama a los métodos de la clase heredada
        self.raza = raza
    def ladrar(self):
        print(f"Guau, mi nombre es {self.nombre}, y soy de raza {self.raza}")

perro1 = Perro(1457,"Sir",1000,"Husky")
perro1.ladrar()
```

Especialización
- Consiste en especificar métodos y/o atributos propios para una clase que recibió herencia.
- En el caso de abajo, la clase gato, heredo los métodos y variables de mascota, pero se agrego territorio
- Para modificar un método heredado, luego de la definición se usa la palabra super. Si se quieren agregar nuevos argumentos se deben insertar después.

```python
class Gato(Mascota):
    def __init__(self, id_chip, nombre, precio, territorio):
        super().__init__(id_chip, nombre, precio)
        self.territorio = territorio
    def maullar(self):
        print(f"Miau, mi nombre es {self.nombre} y mi territorio es {self.territorio}")
gato1= Gato(85,"Botas",1458,"5 cuadras")
gato1.maullar()
```

```python
class Siames(Gato):
    def __init__(self, id_chip, nombre, precio, territorio, numero_manchas_negras):
        super().__init__(id_chip, nombre, precio, territorio)
        self.numero_manchas_negras = numero_manchas_negras
    def maullar(self):
        print(f"Miau, mi nombre es {self.nombre} y mi territorio es {self.territorio}, además tengo {self.numero_manchas_negras} manchas negras.")
gato23= Siames(4552,"Ger",1547,"5 cuadras",35)
gato23.maullar()
```
